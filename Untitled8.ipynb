### **Meaning / Definition of Power BI**

**Power BI** is a **Business Intelligence (BI) and Data Visualization tool** developed by **Microsoft**.
It is used to **collect, analyze, transform, and visualize data** from different sources and convert raw data into **meaningful reports, dashboards, and insights** that help management in **decision-making**.

Power BI allows users to:

* Connect to multiple data sources
* Clean and transform data
* Create interactive reports and dashboards
* Share insights securely across an organization

---

## **Power BI Ecosystem**

The **Power BI ecosystem** consists of various tools and services that work together to **create, publish, share, and access reports**.
The **main components** of the Power BI ecosystem are:

1. Power BI Desktop
2. Power BI Service
3. Power BI Mobile
4. Power BI Gateway

## **1. Power BI Desktop**

### **Meaning**

Power BI Desktop is a **free desktop application** used to **create reports and data models**.

### **Functions**

* Connects to different data sources (Excel, SQL Server, CSV, Web, etc.)
* Cleans and transforms data using **Power Query**
* Creates data models using **DAX (Data Analysis Expressions)**
* Designs interactive **charts, tables, maps, and visuals**

### **Features**

* Drag-and-drop interface
* Advanced data modeling
* Offline report creation

### **Use**

It is mainly used by **data analysts and report developers** to build reports before publishing them.

## **2. Power BI Service**

### **Meaning**

Power BI Service is a **cloud-based online platform** (PowerBI.com) used to **publish, share, and collaborate** on reports.

### **Functions**

* Publish reports created in Power BI Desktop
* Create dashboards from reports
* Share reports with colleagues
* Schedule automatic data refresh

### **Features**

* Cloud storage
* Collaboration and sharing
* Role-based access and security

### **Use**

Used by **organizations and management** to view reports and dashboards from anywhere.


## **3. Power BI Mobile**

### **Meaning**

Power BI Mobile is a **mobile application** available on **Android, iOS, and Windows devices**.

### **Functions**

* View reports and dashboards on mobile phones and tablets
* Receive real-time alerts
* Interact with data using touch features

### **Features**

* User-friendly mobile interface
* Real-time notifications
* Secure access

### **Use**

Helpful for **managers and executives** who need data insights **on the go**.

## **4. Power BI Gateway**

### **Meaning**

Power BI Gateway is a **bridge between on-premises data sources and Power BI Service**.

### **Functions**

* Allows Power BI Service to access local data (SQL Server, Oracle, etc.)
* Enables scheduled data refresh
* Maintains data security

### **Types**

* **Personal Mode** – for individual use
* **Enterprise Mode** – for organization-wide use

### **Use**

Used when data is stored **inside company servers** and not in the cloud.

## **Conclusion**

Power BI is a powerful Business Intelligence tool that helps organizations convert data into insights.
The **Power BI ecosystem**—Desktop, Service, Mobile, and Gateway—works together to provide a **complete end-to-end data analytics solution**, making Power BI widely used in modern business environments.

# **1. Pie Chart vs Donut Chart**

### **Pie Chart**

**Meaning:**
A Pie Chart is a **circular visual** divided into slices, where each slice represents a **part of a whole (100%)**.

### **Features**

* Shows percentage or proportion
* Easy to understand
* Best for small datasets

### **Limitations**

* Difficult to compare many categories
* Not suitable when values are close



### **Donut Chart**

**Meaning:**
A Donut Chart is similar to a Pie Chart but has a **hole in the center**, which can display additional information.

### **Features**

* Better visual appeal
* Can display total or label in the center
* Slightly better readability than pie chart

### **Limitations**

* Still not ideal for many categories
* Can be misleading if overused


### **Comparison: Pie Chart vs Donut Chart**

| Basis         | Pie Chart                 | Donut Chart             |
| ------------- | ------------------------- | ----------------------- |
| Shape         | Full circle               | Circle with center hole |
| Readability   | Good for fewer categories | Better than pie chart   |
| Data Display  | Proportion of whole       | Proportion + extra info |
| Visual Appeal | Simple                    | More attractive         |


### **When to Prefer**

* **Prefer Pie Chart:**
  When you want to show **simple percentage distribution** with very few categories.

* **Prefer Donut Chart:**
  When you want a **modern look** and need to display **total value in the center**.


### **Example**

* **Pie Chart:** Market share of companies (Company A – 40%, B – 30%, C – 30%)
* **Donut Chart:** Department-wise expense contribution with total expenses shown in center



# **2. Bar Chart vs Column Chart**

### **Bar Chart**

**Meaning:**
A Bar Chart displays data using **horizontal bars**.

### **Features**

* Easy to read long category names
* Good for ranking comparison
* Suitable when many categories exist


### **Column Chart**

**Meaning:**
A Column Chart displays data using **vertical bars**.

### **Features**

* Best for time-based data
* Easy to compare changes over periods
* Widely used for trends



### **Comparison: Bar Chart vs Column Chart**

| Basis             | Bar Chart       | Column Chart     |
| ----------------- | --------------- | ---------------- |
| Orientation       | Horizontal bars | Vertical bars    |
| Best for          | Many categories | Time-series data |
| Label readability | Very good       | Limited space    |
| Trend analysis    | Less effective  | Very effective   |



### **When to Prefer**

* **Prefer Bar Chart:**
  When category names are **long** or you need to **rank data clearly**.

* **Prefer Column Chart:**
  When analyzing **data over time** like monthly or yearly performance.

### **Example**

* **Bar Chart:** Sales by product category (Electronics, Furniture, Appliances)
* **Column Chart:** Monthly sales of a company for one year


## **Conclusion**

Choosing the correct Power BI visual improves **clarity, accuracy, and decision-making**.
Pie vs Donut charts are best for **part-to-whole analysis**, while Bar vs Column charts are ideal for **comparative and trend analysis**.

## **1. Star Schema vs Snowflake Schema**

In Power BI, **data modeling** is very important for performance and accurate analysis.
**Star schema** and **Snowflake schema** are two common data modeling techniques used in Power BI.


### **(a) Star Schema**

**Meaning:**
In a Star Schema, there is **one central fact table** connected directly to **multiple dimension tables**.
The structure looks like a **star**.

**Features / Significance:**

* Simple and easy to understand
* Fewer joins → **better performance**
* Faster report and query execution
* Preferred and recommended in Power BI

**Use:**
Used in **most Power BI models** for reporting and dashboards.


### **(b) Snowflake Schema**

**Meaning:**
In a Snowflake Schema, **dimension tables are further split into sub-tables**, forming a complex structure.

**Features / Significance:**

* Saves storage space
* More normalized structure
* Requires multiple joins → **slower performance**
* Complex to understand and maintain

**Use:**
Used when dimensions are **very large and complex**.


### **Comparison: Star Schema vs Snowflake Schema**

| Basis             | Star Schema        | Snowflake Schema       |
| ----------------- | ------------------ | ---------------------- |
| Structure         | Simple star-like   | Complex snowflake-like |
| Tables            | Fewer tables       | More tables            |
| Performance       | High               | Lower                  |
| Joins             | Less               | More                   |
| Usage in Power BI | Highly recommended | Less preferred         |


## **2. Primary Key vs Foreign Key (Power BI Relationships)**

Power BI uses **relationships** to connect tables, similar to databases.


### **(a) Primary Key**

**Meaning:**
A Primary Key is a column that **uniquely identifies each record** in a table.

**Significance:**

* No duplicate values
* Cannot contain NULL values
* Used on the **one side** of relationship

**Example:**
CustomerID in Customer table


### **(b) Foreign Key**

**Meaning:**
A Foreign Key is a column that **refers to the primary key of another table**.

**Significance:**

* Can contain duplicate values
* Used on the **many side** of relationship
* Helps in linking tables

**Example:**
CustomerID in Sales table


### **Comparison: Primary Key vs Foreign Key**

| Basis       | Primary Key     | Foreign Key         |
| ----------- | --------------- | ------------------- |
| Uniqueness  | Unique          | Can repeat          |
| NULL values | Not allowed     | Allowed             |
| Location    | Dimension table | Fact table          |
| Purpose     | Identify record | Create relationship |


## **3. Cardinality in Power BI**

### **Meaning of Cardinality**

**Cardinality** defines the **relationship type between two tables** in Power BI.


### **Types of Cardinality**

* **One-to-Many (1:*)** – Most common
* **Many-to-One (*:1)**
* **One-to-One (1:1)**
* **Many-to-Many (*:*)**


### **Why Cardinality is Important**

Cardinality is important because it:

* Controls **how data flows between tables**
* Ensures **accurate calculations**
* Prevents **wrong totals and duplication**
* Improves **model performance**
* Helps Power BI apply correct filters


### **Example**

* One Customer → Many Sales
  (CustomerID → Sales table)

This relationship must be **One-to-Many** for correct sales totals.


## **Conclusion**

In Power BI:

* **Star schema** is preferred due to simplicity and high performance.
* **Primary and Foreign keys** help create accurate table relationships.
* **Cardinality** ensures correct data filtering and calculations, making it a crucial part of Power BI data modeling.




## **1. Calculated Column vs Measure**

In Power BI, **DAX (Data Analysis Expressions)** is used to create **Calculated Columns** and **Measures**.
Though both use DAX, they serve **different purposes**.



### **(a) Calculated Column**

**Meaning:**
A Calculated Column is a **new column created in a table** using DAX.
It is calculated **row by row** and the result is **stored in the model**.

**Features:**

* Calculated during data refresh
* Uses **row context**
* Increases model size
* Value does not change with filters (except slicers)

**Example:**

```
TotalPrice = Sales[Quantity] * Sales[UnitPrice]
```



### **(b) Measure**

**Meaning:**
A Measure is a **calculation used for analysis**, created using DAX.
It is calculated **only when used in a visual**.

**Features:**

* Calculated at query time
* Uses **filter context**
* Does not increase model size
* Value changes dynamically with filters and slicers

**Example:**

```
Total Sales = SUM(Sales[SalesAmount])
```


### **Difference: Calculated Column vs Measure**

| Basis             | Calculated Column      | Measure              |
| ----------------- | ---------------------- | -------------------- |
| Calculation level | Row level              | Aggregate level      |
| Storage           | Stored in model        | Not stored           |
| Context           | Row context            | Filter context       |
| Usage             | Table visuals, slicing | KPIs, totals, charts |
| Performance       | Lower                  | Better               |


## **2. Row Context**

### **Meaning**

Row context means **Power BI evaluates one row at a time** while calculating a formula.

### **Where it is used**

* Calculated columns
* Iterators (SUMX, AVERAGEX)


### **Simple Example (Row Context)**

Sales Table:

| Quantity | Price |
| -------- | ----- |
| 2        | 100   |
| 3        | 200   |

Calculated Column:

```
Amount = Sales[Quantity] * Sales[Price]
```

👉 Power BI calculates **each row separately**.


## **3. Filter Context**

### **Meaning**

Filter context means **Power BI applies filters** (from slicers, visuals, rows, columns) **before calculating a measure**.

### **Where it is used**

* Measures
* Visuals and dashboards



### **Simple Example (Filter Context)**

Measure:

```
Total Sales = SUM(Sales[Amount])

If a slicer filters **Year = 2024**,
👉 Total Sales is calculated **only for 2024 data**.



## **Row Context vs Filter Context**

| Basis                | Row Context        | Filter Context     |
| -------------------- | ------------------ | ------------------ |
| Applies to           | Single row         | Set of rows        |
| Used in              | Calculated columns | Measures           |
| Changes with slicers | ❌ No               | ✅ Yes              |
| Example              | Price × Quantity   | SUM(Sales) by Year |



## **Conclusion**

* **Calculated Columns** are used for **row-level calculations** and stored in the model.
* **Measures** are used for **dynamic analysis** and respond to filters.
* **Row context** works row by row, while **filter context** works on filtered data sets.



Question 5:
What is the difference between a report and a dashboard in Power BI?




## **1. Report in Power BI**

### **Meaning**

A **Report** is a **detailed, multi-page collection of visuals** created using data from **one or more datasets**.

### **Features**

* Created in **Power BI Desktop**
* Can have **multiple pages**
* Allows **deep data analysis**
* Fully interactive (filters, slicers, drill-down)
* Can be shared through Power BI Service

### **Use**

Used by **analysts and operational users** for detailed analysis.

### **Example**

A sales report with:

* Page 1: Overall sales
* Page 2: Region-wise sales
* Page 3: Product-wise sales


## **2. Dashboard in Power BI**

### **Meaning**

A **Dashboard** is a **single-page summary view** made up of visuals (tiles) pinned from **one or more reports**.

### **Features**

* Created only in **Power BI Service**
* Always **one page**
* Provides **high-level overview**
* Can combine visuals from different datasets
* Less interactive than reports

### **Use**

Used by **managers and executives** for quick decision-making.

### **Example**

A dashboard showing:

* Total sales
* Profit
* Top product
* Sales trend (all on one screen)


## **Difference between Report and Dashboard**

| Basis         | Report            | Dashboard          |
| ------------- | ----------------- | ------------------ |
| Pages         | Multiple pages    | Single page        |
| Creation tool | Power BI Desktop  | Power BI Service   |
| Data source   | One dataset       | Multiple datasets  |
| Detail level  | Detailed analysis | High-level summary |
| Interactivity | High              | Limited            |
| Target users  | Analysts          | Executives         |


## **Conclusion**

In Power BI, **reports** are used for **in-depth data analysis**, while **dashboards** provide a **quick snapshot of key metrics**.
Both together help organizations make **effective and timely decisions**.
